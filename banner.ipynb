```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

# Figures

```python
samples_0 = np.random.beta(a=2,b=12,size=100000)

samples_1 = np.random.beta(a=12,b=2,size=100000)
samples_2 = np.random.beta(a=12,b=2,size=100000)
samples_3 = np.random.beta(a=12,b=2,size=100000)

samples_4 = np.random.beta(a=12,b=6,size=100000)
samples_5 = np.random.beta(a=12,b=6,size=100000)

samples_6 = np.random.beta(a=6,b=10,size=100000)

sns.set(rc={'axes.facecolor': 'white', 'figure.facecolor': 'white'})

sns.histplot(samples_0)
sns.histplot(samples_1)
sns.histplot(samples_2)
sns.histplot(samples_3)
sns.histplot(samples_4)
sns.histplot(samples_5)
sns.histplot(samples_6)

plt.grid(False)
```

```python
samples = np.random.beta(a=12, b=2, size=100000)

sns.set(rc={'axes.facecolor': 'white', 'figure.facecolor': 'white'})

sns.histplot(samples, color="#0f4461")

plt.grid(False)
```
